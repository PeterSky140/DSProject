```python
df_na=df_all.isna().sum()/len(df_all) * 100 # ou df_all.isna().sum()/df_all.shape[0] * 100
```

```python
df_na_sorted=df_na.sort_values(ascending=False)
df_na_sorted[df_na_sorted>10]
```

```python
df_na_sorted=df_na_sorted.reset_index()
df_na_sorted.columns=['Variable', 'Percentage of Missing Values']
```

```python
df_na_sorted
```

```python
threshold=[20,30,40,50,60,70,80,90]
for i in threshold:
    conta=0
    for j in df_na_sorted['Percentage of Missing Values']:
        if j>i:
            conta+=1
    print(f"Number of variables with missing values greater than {i}%: {conta}")
```

Using KNN Imputer to fill the Missing Values from the Numerical Features

```python
from sklearn.impute import KNNImputer
```

```python
num_cols = df_all1.select_dtypes(include="number").columns
imputer = KNNImputer(n_neighbors=5)
```

```python
df_all1[num_cols]
```

```python
df_all1[num_cols] = imputer.fit_transform(df_all1[num_cols])
```
